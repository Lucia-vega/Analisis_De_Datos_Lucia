#  <span style = "color: plum"> Metodos iloc y loc </span>
1. ¿Cuál elegir?
Usa loc: Cuando conoces los nombres (etiquetas) de tus filas o columnas. Es el más humano y fácil de leer.

Usa iloc: Cuando conoces la posición (el orden numérico). Es ideal para cuando quieres "las primeras 5 filas" o "la última columna".

2. La regla del Rango (Abierto vs. Cerrado)
Esta es la parte donde la mayoría se equivoca. Memoriza esto:

loc es INCLUSIVO (Cerrado): Se incluye tanto el inicio como el final.

40:50 te da 11 filas (de la 40 a la 50 inclusive).

iloc es EXCLUSIVO (Abierto): Como las listas de Python, el final no se incluye.

40:50 te da 10 filas (de la 40 a la 49).

3. Ejemplos Prácticos
Imagina que tu columna se llama "edad" y es la columna número 3 (índice 2).

Por Nombre (loc)
Comando: df.loc[40:50, 'edad']

Filas: Trae desde la etiqueta 40 hasta la 50.

Columna: Busca específicamente la palabra 'edad'.

Por Posición (iloc)
Comando: df.iloc[40:51, 2]

Filas: Trae desde la posición 40 hasta la 50 (ponemos 51 para que incluya la 50).

Columna: Trae la columna en la posición 2 (la tercera).
