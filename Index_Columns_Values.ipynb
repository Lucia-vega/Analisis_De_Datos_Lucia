# 📌 Recordatorio: Diferencias de Formato en Pandas

### 1. `df["Columna"]` (Pandas Series)
Es el formato estándar de Pandas. Incluye los datos y el **índice**.
* **Cuándo usar:** Para limpieza de datos, filtros, cálculos estadísticos (`.mean()`, `.sum()`) y visualización.
* **Ventaja:** Mantiene la relación con las etiquetas del DataFrame.

### 2. `df["Columna"].values` (NumPy Array)
Extrae los datos puros en forma de matriz (o vector), eliminando el índice.
* **Cuándo usar:** * Para alimentar modelos de **Machine Learning** (Scikit-Learn).
    * Para operaciones matemáticas de alto rendimiento con **NumPy**.
* **Ventaja:** Es más ligero y es el formato "universal" para cálculos científicos en Python.
