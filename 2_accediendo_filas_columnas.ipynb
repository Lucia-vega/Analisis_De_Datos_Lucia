## **Métodos `iloc` y `loc`**

Son dos de los `métodos` más importantes en **Pandas** para acceder a datos dentro de un `DataFrame`. Aunque son similares, tienen diferencias clave en cómo seleccionan filas y columnas.

| Método | Tipo índice | Qué acepta | Ejemplo |
| --- | --- | --- | --- |
| `iloc[fila, columna]` | Índices enteros (posición) | enteros, listas, rangos, slices | `df.iloc[0, 1]` |
| `loc[fila, columna]` | Índices etiquetados (nombre) | etiquetas, listas de etiquetas, condiciones booleanas | `df.loc["Juan", "edad"]` |


*** El método `iloc` aparece como ***deprecated*** en la documentación de **pandas**.

# 🛠️ El Parámetro `inplace=True` en Pandas

Por defecto, la mayoría de las operaciones en Pandas (como `set_index`, `dropna`, o `rename`) **no modifican** el DataFrame original; en su lugar, devuelven una **copia** con los cambios aplicados.

### ¿Qué hace específicamente?
Cuando activas `inplace=True`, le das una instrucción directa a Pandas:
* **Modificación directa:** Los cambios se aplican sobre el objeto original (`df_animales`).
* **Sin retorno:** El método devuelve `None`. Por eso, si haces `df = df.set_index(..., inplace=True)`, tu variable `df` quedará vacía.

### Comparativa de uso

| Sin `inplace` (Recomendado) | Con `inplace=True` |
| :--- | :--- |
| `df = df.set_index("animal")` | `df.set_index("animal", inplace=True)` |
| Crea una copia y la reasigna. | Modifica el DataFrame existente. |
| Permite "encadenar" métodos (pipeline). | No permite encadenar más funciones. |

---

### ⚠️ Nota de Buenas Prácticas
Aunque es muy común encontrarlo en tutoriales, la comunidad de Data Science y los desarrolladores de Pandas recomiendan **no usarlo** por las siguientes razones:

1. **Legibilidad:** Es más claro ver `df = df.operacion()` para saber que la variable está cambiando.
2. **Consistencia:** No todos los métodos de Pandas tienen el parámetro `inplace`, lo que genera confusión.
3. **Depreciación:** Pandas está trabajando para eliminar este parámetro en versiones futuras (Pandas 3.0+), por lo que es mejor acostumbrarse a la reasignación.



### Ejemplo en el Notebook:
```python
# ✅ MEJOR PRÁCTICA (Nada de inplace=True)
df_animales = df_animales.set_index("animal")

# ❌ FORMA ANTIGUA
# df_animales.set_index("animal", inplace=True)

# 📉 ¿Qué pasa si no reasigno ni uso `inplace=True`?

Si ejecutas una operación como `df.set_index("animal")` por sí sola, sucede lo siguiente:

1. **Se crea una copia temporal:** Pandas procesa la instrucción y genera un nuevo DataFrame en la memoria RAM con el cambio aplicado.
2. **Se muestra en pantalla:** Como no guardaste ese resultado en ninguna parte, el Notebook simplemente te muestra una "vista previa" del resultado en la celda.
3. **El cambio se descarta:** Al terminar de ejecutarse la celda, esa copia temporal desaparece. Tu DataFrame original `df_animales` **sigue igual que al principio**.

### 🧪 Ejemplo del "Efecto Fantasma"

```python
# Escenario: Queremos cambiar el índice a la columna 'animal'

# ❌ ERROR COMÚN: La operación se ejecuta, se muestra, pero NO se guarda.
df_animales.set_index("animal") 

# Si revisas el DF en la siguiente celda, verás que NO tiene el nuevo índice.
print(df_animales.index)
