# <span style = "color: blue"> Copiando DataFrame </span>

### 📝 Resumen: ¿Por qué usar `.copy()` en Pandas?

Usar `df = df_original.copy()` es una **regla de oro** en el análisis de datos por las siguientes razones:



#### 1. Independencia Total (Evita el "Efecto Espejo")
* **Sin `.copy()`:** Si haces `df = df_original`, ambos nombres apuntan a los **mismos datos**. Lo que borres o cambies en uno, afectará automáticamente al otro.
* **Con `.copy()`:** Creas un objeto nuevo e independiente. Si "rompes" el nuevo `df`, tus datos originales permanecen a salvo.

#### 2. Adiós al `SettingWithCopyWarning`
Pandas lanza este aviso cuando no está seguro si estás modificando una copia o el original. Al usar `.copy()`, le confirmas a Python que el objeto es independiente, eliminando avisos confusos.

#### 3. Flujo de Trabajo Seguro
Mantiene los datos "crudos" en memoria. Si te equivocas limpiando los datos, no necesitas volver a cargar el archivo (CSV/Excel) desde el disco; simplemente vuelves a ejecutar la copia y empiezas de nuevo.

---

**Ejemplo de uso:**
```python
# Mantiene el original intacto para emergencias
df_limpio = df_original.copy() 

# Ahora puedes trabajar sin miedo
df_limpio.dropna(inplace=True)

````python
df = df_Original.copy()
````

### 🗑️ Método `.drop()` en Pandas

El método **`.drop()`** se utiliza para eliminar filas o columnas de un DataFrame.

#### Desglose del comando:
`df.drop(columns=["nombre"], inplace=True)`

* **`columns=["nombre"]`**: Especifica la columna (o lista de columnas) que deseas eliminar.
* **`inplace=True`**: Modifica el DataFrame original directamente. Si no se usa, Pandas genera una copia con el cambio pero no altera el original.

---
**Diferencia entre Axis:**
* Aunque usar `columns=` es más moderno, a veces verás `axis=1`. Ambos significan lo mismo:
    * `axis=1` o `columns`: Se refiere a **Columnas** (vertical).
    * `axis=0` o `index`: Se refiere a **Filas** (horizontal).

**Ejemplo:**
```python
# Borra la columna 'comision' definitivamente
df_Original.drop(columns=["comision"], inplace=True)

### 🏷️ Renombrar columnas masivamente

El comando `df.columns = lista_de_nombres` se utiliza para asignar nuevos títulos a todas las columnas de un DataFrame de forma simultánea.

#### Características clave:
* **Orden posicional:** El primer nombre de la lista se asignará a la primera columna, el segundo a la segunda, etc.
* **Requisito de longitud:** La lista debe tener el mismo número de elementos que el DataFrame original.
* **Rapidez:** Es ideal cuando los nombres originales vienen mal (por ejemplo, con espacios o caracteres extraños) y quieres limpiarlos todos de una vez.

---
**Ejemplo de uso:**
```python
# Definimos los nuevos nombres
nuevos_nombres = ['fecha', 'producto', 'cantidad', 'precio']

# Los aplicamos al DataFrame
df.columns = nuevos_nombres

### 🔢 Conversión de tipos con `.astype()`

El método **`.astype()`** se usa para cambiar el tipo de dato de una columna (por ejemplo, de número a texto).

#### Caso común: Convertir a String (`str`)
`df["columna"] = df["columna"].astype("str")`

* **¿Para qué sirve?**: Muy útil para columnas que contienen IDs, códigos postales, números de teléfono o matrículas. 
* **Importancia**: Evita que se pierdan ceros iniciales (ej: "0800") y previene errores al intentar realizar operaciones matemáticas en datos que solo sirven como identificadores.
* **Nota**: Debes reasignar la columna (`df["col"] = ...`) para que el cambio sea permanente en el DataFrame.
