## <span style="color: plum">¿Qué es la Covarianza? 📊</span>

La **Covarianza** es una medida estadística que indica la dirección de la relación lineal entre dos variables aleatorias. En términos simples, nos dice si dos variables "se mueven juntas" o en direcciones opuestas.

### 🧠 Interpretación de los Resultados

Para analizar nuestra matriz, debemos observar dos factores clave:

1. **El Signo (Dirección):**
    * **Positivo (+):** Las variables tienen una relación directa. Si una aumenta, la otra también (ej. *Precio Costo* y *Valor Final*).
    * **Negativo (-):** Las variables tienen una relación inversa. Si una aumenta, la otra disminuye (ej. *Precio* y *Cantidad*).
    * **Cero (0):** No existe una relación lineal aparente entre las variables.

2. **La Magnitud (Varianza):**
    * Los valores en la **diagonal principal** representan la **Varianza** de cada variable (qué tanto se alejan los datos de su propio promedio).



---

### 🔍 Análisis de mis Datos (Hallazgos Relevantes)

| Relación | Tipo de Covarianza | Significado en el Negocio |
| :--- | :--- | :--- |
| **Precios vs. Valor Final** | **Positiva Alta** | El modelo de precios es consistente; el aumento en costos se refleja directamente en el ingreso final. |
| **Cantidad vs. Valor Unitario** | **Negativa** | Existe una ligera tendencia donde los productos de mayor valor se venden en menores cantidades (Elasticidad). |
| **Diagonal Principal** | **Varianza** | Nos indica qué tan dispersos son los valores de cada columna por sí sola. |

> **Nota importante:** La covarianza depende de las unidades de medida (pesos vs. unidades), por lo que números muy grandes no siempre significan una relación más fuerte que los números pequeños.

## <span style="color: plum">¿Qué es la Correlación? 📈</span>

La **Correlación (de Pearson)** es una versión normalizada de la covarianza. Nos indica no solo la dirección de la relación, sino también la **fuerza** de la misma, eliminando el problema de las unidades de medida (pesos vs. unidades).

### 🧠 Escala de Interpretación

A diferencia de la covarianza, aquí los números siempre están entre **-1 y 1**:

* **1.0:** Correlación positiva perfecta (suben exactamente igual).
* **0.8 a 0.99:** Correlación muy fuerte.
* **0.0:** No hay relación lineal.
* **-1.0:** Correlación negativa perfecta (una sube y la otra baja proporcionalmente).


---

### 🔍 Análisis de mis Resultados (Correlación)

| Relación | Coeficiente | Interpretación Técnica |
| :--- | :--- | :--- |
| **Costo vs. Valor Unitario** | **0.99** | **Relación Casi Perfecta.** El valor unitario depende casi totalmente del costo de adquisición. |
| **Costo/Unitario vs. Valor Final** | **0.81 / 0.82** | **Relación Muy Fuerte.** El precio que paga el cliente está fuertemente ligado a los costos. |
| **Cantidad vs. Precios** | **-0.04** | **Relación Nula/Inexistente.** El precio del producto no parece afectar cuántas unidades se compran (comportamiento inelástico). |
| **Cantidad vs. Valor Final** | **0.39** | **Relación Débil.** El total de la venta sube un poco cuando hay más cantidad, pero los precios influyen más. |

### 💡 Conclusión Clave
Mientras que la covarianza nos daba números de millones que eran difíciles de comparar, la **correlación** nos confirma que la relación más sólida en tu dataset es entre el **costo y el valor unitario (0.99)**, y que la **cantidad** es la variable más independiente del grupo.

## <span style="color: plum">Operadores Lógicos 🧠</span>

Los operadores lógicos nos permiten combinar múltiples comparaciones para crear filtros más complejos. Son la base para tomar decisiones en el código.

### 📋 Tabla de Operadores Lógicos

| Operador | Nombre | Descripción | Ejemplo |
| :--- | :--- | :--- | :--- |
| **and** (&) | **Y lógico** | Devuelve `True` si **ambos** operadores son `True`. | `True and False` -> **False** |
| **or** (\|) | **O lógico** | Devuelve `True` si **al menos uno** de los operadores es `True`. | `True or False` -> **True** |

* Usa **`~`** en lugar de `not`.

## <span style="color: plum">Concepto de Slicing (Rebanado) en Pandas 🍞</span>

El **Slicing** es una técnica que nos permite extraer una "rebanada" o un rango específico de datos de nuestro DataFrame. En lugar de seleccionar elementos uno por uno, definimos un **punto de inicio** y un **punto de fin**.

### 🔍 Anatomía de un Slice
La sintaxis básica utiliza los dos puntos `:` para separar los límites:

> **`rango = [ inicio : fin : paso ]`**

* **Inicio:** Desde dónde empezamos (incluido).
* **Fin:** Hasta dónde llegamos (en etiquetas de Pandas, se incluye el final).
* **Paso:** De cuánto en cuánto saltamos (por defecto es 1).



---

### 🛠️ Análisis de mi código con `.loc`

En la línea que ejecutamos:
`df.loc[(df["codigo_producto"] == "10015") | (df["codigo_producto"] == "10003"), "codigo_producto":]`

#### 1. Filtro de Filas (Antes de la coma)
Usamos el operador OR (`|`) para traer solo los productos con códigos **10015** o **10003**.

#### 2. Slicing de Columnas (Después de la coma)
Al escribir `"codigo_producto":`, aplicamos un slice que dice:
* **Inicio:** Columna `codigo_producto`.
* **Fin:** (Vacío), lo que significa "hasta el final de la tabla".

**Resultado:** El DataFrame nos muestra todas las características de esos productos, pero **oculta** las columnas que estén a la izquierda del código (como IDs internos o fechas que no necesitamos ver en ese momento).

---

### 💡 Atajos rápidos de Slicing

| Código | Significado |
| :--- | :--- |
| `[:]` | Selecciona **todo** el eje (filas o columnas). |
| `[:"precio"]` | Desde el **principio** hasta la columna "precio". |
| `"cantidad":` | Desde "cantidad" hasta el **final**. |
| `::2` | Selecciona uno de cada dos (salta de **2 en 2**). |

## <span style="color: plum">Aplicando Queries ⚡</span>

El método `.query()` es una herramienta de Pandas que nos permite consultar y filtrar columnas utilizando **expresiones booleanas** de una manera mucho más legible, similar al lenguaje natural o a SQL.

### 📝 Sintaxis Básica
A diferencia de `.loc[]`, en una query pasamos toda la condición como una cadena de texto (string).

```python
# Ejemplo: Filtrar productos con valor final mayor a 1000
df.query("valor_final > 1000")

````python
df["categoria"].unique()
````

### 🔑 Regla de las Comillas en `.query()`

Para filtrar por texto (strings) dentro de una query, debemos usar **comillas anidadas**:

1. **Comillas Externas:** Delimitan toda la consulta de Pandas.
2. **Comillas Internas:** Delimitan el valor específico que estamos buscando dentro de la columna.

> **Regla de oro:** Si usas comillas dobles `"` para la consulta, usa comillas sencillas `'` para el texto interno (o viceversa). 

* **Correcto:** `df.query("color == 'Rojo'")`
* **Incorrecto:** `df.query("color == Rojo")` -> (Pandas buscará una columna llamada Rojo).

````python
df.query("categoria == 'Electrónicos'")
````

````python
df.query("categoria == 'Electrónicos' and matricula_funcionario == '123456999'")
````

## <span style="color: plum">¿Cuándo usar `()` o `[]`? 🧩</span>

| Símbolo | Nombre | Cuándo usarlo | Ejemplo |
| :--- | :--- | :--- | :--- |
| **`( )`** | **Paréntesis** | Para ejecutar **acciones** (métodos). Van después de verbos. | `.sum()`, `.groupby()`, `.head()` |
| **`[ ]`** | **Corchetes** | Para **seleccionar** columnas o filtrar datos. | `df["columna"]`, `df[df["A"] > 1]` |
| **`[[ ]]`**| **Doble Corchete**| Para seleccionar columnas y mantener el formato de **tabla (DataFrame)**. | `df[["precio", "cantidad"]]` |

> **Regla rápida:** Si le pides a Pandas que **HAGA** algo, usa `()`. Si le pides que **BUSQUE** algo, usa `[]`.

## <span style="color: plum">El Poder de `.agg()` (Agregaciones) 📊</span>

El método `.agg()` nos permite realizar múltiples cálculos matemáticos al mismo tiempo después de un `groupby`.

### 1. Agregación por Diccionario `{ }`
Se usa para aplicar **funciones específicas a columnas específicas**.
* **Estructura:** `{"columna": "operacion"}`
* **Ejemplo:** `df.agg({"costo": "sum", "stock": "mean"})`

### 2. Agregación por Lista `[ ]`
Se usa para aplicar **varias funciones a una sola columna** seleccionada.
* **Estructura:** `["columna"].agg(["op1", "op2"])`
* **Ejemplo:** `df["precio"].agg(["min", "max", "std"])`

---

### 💡 Tip Profesional: Renombrar Columnas
Si quieres que el resultado no diga "sum" o "mean" sino algo más claro, puedes hacer esto:
`df.groupby("vendedor").agg(Ventas_Totales=("cantidad", "sum"))`
