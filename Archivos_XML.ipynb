3. HTML | pd.read_html()
HTML significa HyperText Markup Language (Lenguaje de Marcado de Hipertexto) y es el lenguaje estándar que se utiliza para crear páginas web. El tipo de archivos HMTL está dado por la extensión .html.
El método pd.read_html() permite leer páginas web (usando los protocolos http, ftp y file de URL) en busca de tablas HTML, que son retornadas en una lista de objetos DataFrame. Algunos argumentos de interés son:

Cadena de texto o expresión regular a buscar en el contenido de las tablas. Si se especifica, solo las tablas que cumplan con esta condición serán cargadas. (match)

Etiquetas o atributos HTML usados para identificar una tabla en particular. (attrs)

Consultar la especificación de la función en la documentación oficial para más detalles sobre cómo usar estos argumentos.

Cuando se quiere extraer datos de un sitio de internet, utilizamos una técnica llamada web scraping.
Algunas limitaciones de este método son:

Busca únicamente datos en formato tabular.

En muchas ocasiones es necesario buscar en el codigo HTML con otras bibliotecas que simulen un navegador (requests), porque el sitio web puede bloquear la consulta directa con Pandas.
